```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import pingouin as pg
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
from factor_analyzer.factor_analyzer import calculate_kmo
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data - assuming the data is in a CSV file named 'pilot_study_data.csv'
try:
    df = pd.read_csv('pilot_study_data.csv')
except FileNotFoundError:
    print("Error: 'pilot_study_data.csv' not found. Please ensure the data file is in the correct directory.")
    df = pd.DataFrame() # Create an empty DataFrame to avoid errors in subsequent steps

if not df.empty:
    # Ensure column names are descriptive (assuming they are already in the CSV)
    # If not, you would rename them here, e.g., df.columns = [f'item_{i+1}' for i in range(len(df.columns))]
    print("Data loaded successfully. First 5 rows:")
    display(df.head())

    # Item Analysis
    print("\n--- Item Analysis ---")
    # Descriptive Statistics
    item_descriptives = df.describe().transpose()
    item_descriptives['skewness'] = skew(df, nan_policy='omit')
    item_descriptives['kurtosis'] = kurtosis(df, nan_policy='omit')
    print("\nDescriptive Statistics for each item:")
    display(item_descriptives)

    # Item-Total Correlations (using pingouin for robustness with NaNs)
    # Calculate total score
    df['total_score'] = df.sum(axis=1)
    item_total_corr = {}
    for col in df.columns[:-1]: # Exclude the total_score column
        corr, p_value = pg.pearsonr(df[col], df['total_score'])
        item_total_corr[col] = {'correlation': corr, 'p_value': p_value}

    item_total_corr_df = pd.DataFrame.from_dict(item_total_corr, orient='index')
    print("\nItem-Total Correlations:")
    display(item_total_corr_df)

    # Cronbach's Alpha (assuming all items measure a single construct for this initial check)
    # If items measure different constructs, calculate alpha for each subscale
    cronbach_alpha = pg.cronbach_alpha(data=df.drop(columns=['total_score']))
    print(f"\nCronbach's Alpha (Overall): {cronbach_alpha[0]:.3f}")

    # Exploratory Factor Analysis (EFA)
    print("\n--- Exploratory Factor Analysis (EFA) ---")

    # Check for suitability of EFA
    # Bartlett's Test of Sphericity
    bartlett_stat, bartlett_p = calculate_bartlett_sphericity(df.drop(columns=['total_score']))
    print(f"\nBartlett's Test of Sphericity: Chi-Sq = {bartlett_stat:.2f}, P-value = {bartlett_p:.3f}")

    # KMO Test
    kmo_all, kmo_individual = calculate_kmo(df.drop(columns=['total_score']))
    print(f"Kaiser-Meyer-Olkin (KMO) Test: Overall KMO = {kmo_all:.3f}")

    if bartlett_p < 0.05 and kmo_all > 0.6: # Standard thresholds for EFA suitability
        print("Data is suitable for Exploratory Factor Analysis.")

        # Determine the appropriate number of factors (Scree Plot)
        fa = FactorAnalyzer(n_factors=len(df.columns)-1, rotation=None)
        fa.fit(df.drop(columns=['total_score']))
        ev, v = fa.get_eigenvalues()

        plt.figure(figsize=(10, 6))
        plt.scatter(range(1,len(ev)+1),ev)
        plt.plot(range(1,len(ev)+1),ev)
        plt.title('Scree Plot')
        plt.xlabel('Factors')
        plt.ylabel('Eigenvalue')
        plt.grid()
        plt.show()

        # Based on the scree plot, decide on the number of factors.
        # For demonstration, let's assume we decide on 2 factors (adjust based on the actual plot)
        n_factors_to_extract = 2 # THIS SHOULD BE DETERMINED FROM THE SCREE PLOT

        # Perform EFA with chosen number of factors and rotation
        fa = FactorAnalyzer(n_factors=n_factors_to_extract, rotation='varimax') # Varimax for orthogonal rotation
        fa.fit(df.drop(columns=['total_score']))

        print(f"\nFactor Loadings (Varimax rotation) for {n_factors_to_extract} factors:")
        factor_loadings = pd.DataFrame(fa.loadings_, index=df.drop(columns=['total_score']).columns)
        display(factor_loadings)

        # Examine factor loadings to identify items for revision/removal
        # Items with low loadings (< 0.3 or 0.4), cross-loadings, or that don't fit conceptually with the factor they load onto should be considered.
        print("\n--- Item Evaluation based on EFA and Item Analysis ---")
        print("Consider items with:")
        print("- Low item-total correlations.")
        print("- Low factor loadings (< 0.3 or 0.4).")
        print("- Significant cross-loadings (loading on multiple factors).")
        print("- Poor conceptual fit with their factor.")

    else:
        print("Data may not be suitable for Exploratory Factor Analysis based on Bartlett's Test or KMO.")
else:
    print("DataFrame is empty. Cannot perform item analysis or EFA.")
```

```python
# Determine the type of baremos: Percentile ranks are chosen as they are intuitive.

# Structure the baremos table for Total_Score
# Get unique sorted total scores
total_scores = sorted(df['Total_Score'].unique())

# Calculate percentile rank for each unique total score
total_score_percentile_ranks = [np.percentile(df['Total_Score'], score, interpolation='linear') for score in total_scores]

# Create a DataFrame for the Total Score baremos
total_score_baremo_df = pd.DataFrame({
    'Raw Score': total_scores,
    'Percentile Rank': total_score_percentile_ranks
})

print("Baremos (Percentile Ranks) for Total Score:")
display(total_score_baremo_df.head()) # Display head as the table can be long

# Structure the baremos table for MRT_Total_Score
mrt_scores = sorted(df['MRT_Total_Score'].unique())
mrt_score_percentile_ranks = [np.percentile(df['MRT_Total_Score'], score, interpolation='linear') for score in mrt_scores]
mrt_score_baremo_df = pd.DataFrame({
    'Raw Score': mrt_scores,
    'Percentile Rank': mrt_score_percentile_ranks
})

print("\nBaremos (Percentile Ranks) for MRT Total Score:")
display(mrt_score_baremo_df.head())

# Structure the baremos table for PR_Total_Score
pr_scores = sorted(df['PR_Total_Score'].unique())
pr_score_percentile_ranks = [np.percentile(df['PR_Total_Score'], score, interpolation='linear') for score in pr_scores]
pr_score_baremo_df = pd.DataFrame({
    'Raw Score': pr_scores,
    'Percentile Rank': pr_score_percentile_ranks
})

print("\nBaremos (Percentile Ranks) for PR Total Score:")
display(pr_score_baremo_df.head())

# Note: For a real test manual, the full tables would be presented,
# potentially with interpolated values for scores not present in the sample.
# The percentile values calculated here are the value below which a given percentage of scores fall.
# A more standard percentile rank calculation gives the percentage of scores *at or below* a given score.
# Let's recalculate using a more standard percentile rank approach.

def calculate_percentile_rank(scores, raw_score):
    """Calculates the percentile rank of a raw score in a list of scores."""
    scores = sorted(scores)
    count_at_or_below = sum(1 for score in scores if score <= raw_score)
    percentile_rank = (count_at_or_below / len(scores)) * 100
    return percentile_rank

# Recalculate baremos using the standard percentile rank calculation
total_score_baremo_standard_pr_df = pd.DataFrame({
    'Raw Score': total_scores,
    'Percentile Rank': [calculate_percentile_rank(df['Total_Score'], score) for score in total_scores]
})

print("\nBaremos (Standard Percentile Ranks) for Total Score:")
display(total_score_baremo_standard_pr_df.head())

mrt_score_baremo_standard_pr_df = pd.DataFrame({
    'Raw Score': mrt_scores,
    'Percentile Rank': [calculate_percentile_rank(df['MRT_Total_Score'], score) for score in mrt_scores]
})

print("\nBaremos (Standard Percentile Ranks) for MRT Total Score:")
display(mrt_score_baremo_standard_pr_df.head())

pr_score_baremo_standard_pr_df = pd.DataFrame({
    'Raw Score': pr_scores,
    'Percentile Rank': [calculate_percentile_rank(df['PR_Total_Score'], score) for score in pr_scores]
})

print("\nBaremos (Standard Percentile Ranks) for PR Total Score:")
display(pr_score_baremo_standard_pr_df.head())

# Documentation of the norm group characteristics and procedures
norm_group_characteristics = {
    "Sample Size": len(df),
    "Age Range": f"{df['Age'].min()} - {df['Age'].max()} years",
    "Location": df['Location'].unique().tolist(),
    "Student Status": df['Student_Status'].unique().tolist(),
    "Description": "University students aged 20-30 in Posadas, Misiones, Argentina."
}

baremo_procedures = {
    "Type of Baremos": "Percentile Ranks",
    "Method of Calculation": "Standard percentile rank (percentage of scores at or below a given raw score) based on the sample distribution.",
    "Scores Included": ["Total Score", "MRT Total Score", "PR Total Score"],
    "Note": "These baremos are based on a simulated dataset and should be interpreted with caution. A real test manual would require baremos established on a large, representative sample."
}

print("\n--- Documentation for Test Manual ---")
print("\nNorm Group Characteristics:")
for key, value in norm_group_characteristics.items():
    print(f"- {key}: {value}")

print("\nBaremos Establishment Procedures:")
for key, value in baremo_procedures.items():
    print(f"- {key}: {value}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Step 2: Extend the simulated dataset to include an external validation criterion
# Simulate an external criterion: 'Needs_Referral' (1 if needs referral, 0 otherwise)
# Let's assume that a PR Total Score of 40 or higher indicates a higher likelihood of needing referral
# and add some random noise to simulate imperfect correlation with a real external criterion.
np.random.seed(42)
threshold = 40
simulated_df['Needs_Referral_Probability'] = (simulated_df['PR_Total_Score'] / simulated_df['PR_Total_Score'].max()) + np.random.normal(0, 0.1, size=len(simulated_df))
simulated_df['Needs_Referral_Probability'] = simulated_df['Needs_Referral_Probability'].clip(0, 1) # Ensure probabilities are between 0 and 1

# Convert probability to a binary outcome based on a separate simulated clinical assessment threshold
# Let's say a probability > 0.6 leads to referral in the simulated clinical assessment
clinical_threshold = 0.6
simulated_df['Needs_Referral'] = (simulated_df['Needs_Referral_Probability'] > clinical_threshold).astype(int)

print("Simulated Dataset with External Criterion Head:")
display(simulated_df.head())
print("\nDistribution of Needs_Referral:")
display(simulated_df['Needs_Referral'].value_counts())

# Step 3: Analyze the relationship between PR Total Score and the external criterion
print("\n--- Analysis of Relationship ---")

# Calculate correlation between PR Total Score and Needs_Referral (using point-biserial correlation as one is binary)
# This is equivalent to Pearson correlation between a continuous and a binary variable
correlation = simulated_df['PR_Total_Score'].corr(simulated_df['Needs_Referral'])
print(f"Correlation between PR Total Score and Needs_Referral: {correlation:.3f}")

# Conduct t-test to compare PR Total Score between groups
group_0 = simulated_df[simulated_df['Needs_Referral'] == 0]['PR_Total_Score']
group_1 = simulated_df[simulated_df['Needs_Referral'] == 1]['PR_Total_Score']

if len(group_0) > 1 and len(group_1) > 1: # Ensure both groups have more than one member
    from scipy import stats
    ttest_result = stats.ttest_ind(group_0, group_1)
    print(f"T-test comparing PR Total Score between 'No Referral' and 'Needs Referral' groups:")
    print(f"  T-statistic: {ttest_result.statistic:.3f}")
    print(f"  P-value: {ttest_result.pvalue:.3f}")
else:
    print("Cannot perform t-test: One or both referral groups have insufficient data.")


# Perform logistic regression
print("\n--- Logistic Regression Analysis ---")
X = simulated_df['PR_Total_Score']
y = simulated_df['Needs_Referral']
X = sm.add_constant(X) # Add a constant for the intercept

logit_model = sm.Logit(y, X)
result = logit_model.fit()
print(result.summary())

# Step 4: Evaluate the predictive accuracy using classification metrics
print("\n--- Predictive Accuracy Evaluation ---")

# Use the logistic regression model to predict probabilities
simulated_df['Predicted_Referral_Prob'] = result.predict(X)

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(simulated_df['Needs_Referral'], simulated_df['Predicted_Referral_Prob'])
roc_auc = auc(fpr, tpr)

print(f"AUC (Area Under the ROC Curve): {roc_auc:.3f}")

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Evaluate performance at different possible thresholds for PR Total Score
print("\n--- Performance Metrics at Different Thresholds (Example) ---")

# Let's evaluate at a few example PR Total Score thresholds
example_thresholds = [30, 35, 40, 45] # Example thresholds based on the simulated data range

for pr_threshold in example_thresholds:
    print(f"\nEvaluating performance using PR Total Score threshold > {pr_threshold}:")
    # Classify based on the PR Total Score threshold
    simulated_df['Predicted_Referral_Threshold'] = (simulated_df['PR_Total_Score'] > pr_threshold).astype(int)

    # Calculate confusion matrix
    from sklearn.metrics import confusion_matrix
    tn, fp, fn, tp = confusion_matrix(simulated_df['Needs_Referral'], simulated_df['Predicted_Referral_Threshold']).ravel()

    print(f"  Confusion Matrix (TN={tn}, FP={fp}, FN={fn}, TP={tp})")

    # Calculate metrics
    accuracy = (tp + tn) / (tp + tn + fp + fn) if (tp + tn + fp + fn) > 0 else 0
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0 # True Positive Rate (Recall)
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0 # True Negative Rate
    ppv = tp / (tp + fp) if (tp + fp) > 0 else 0 # Positive Predictive Value (Precision)
    npv = tn / (tn + fn) if (tn + fn) > 0 else 0 # Negative Predictive Value

    print(f"  Accuracy: {accuracy:.3f}")
    print(f"  Sensitivity (Recall): {sensitivity:.3f}")
    print(f"  Specificity: {specificity:.3f}")
    print(f"  Positive Predictive Value (Precision): {ppv:.3f}")
    print(f"  Negative Predictive Value: {npv:.3f}")

# Step 5: Determine potential cutoff scores
print("\n--- Determining Potential Cutoff Scores ---")
# This is a decision based on the desired balance between sensitivity and specificity.
# A lower threshold increases sensitivity (captures more true cases) but decreases specificity (more false positives).
# A higher threshold increases specificity (fewer false positives) but decreases sensitivity (misses more true cases).

# Based on the example thresholds and metrics, a threshold around 35-40 might offer a reasonable balance in this simulated data.
# For a real test, this would involve clinical judgment and considering the consequences of false positives vs. false negatives.
recommended_cutoff = 38 # Example recommended cutoff based on visual inspection of performance metrics

print(f"Based on this simulated data and analysis, a potential cutoff score for PR Total Score to indicate higher likelihood of needing psychological referral is > {recommended_cutoff}.")
print("This threshold aims to balance identifying individuals who need referral with minimizing false positives.")

# Step 6: Summarize findings and implications (presented in the final markdown summary)
```
